# Google Search Analysis with Python

The pytrends library provides an interface to the Google Trends API, which allows us to retrieve data on the popularity of search terms over time. This can be useful for performing Google search analysis and identifying trends and patterns in search behavior.

To get started with pytrends, we first need to install the library using pip:

Once installed, we can use the pytrends.request.TrendReq() method to create an instance of the TrendReq class, which we can use to send requests to the Google Trends API.

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
# Import the TrendReq class from the pytrends library
from pytrends.request import TrendReq

# Create a TrendReq object
trends = TrendReq()

# Build the payload with the search term "Machine Learning"
trends.build_payload(kw_list=["Machine Learning"])

# Retrieve the interest by region data
data = trends.interest_by_region()

# Sort the data by the "Machine Learning" column in descending order
data = data.sort_values(by="Machine Learning", ascending=False)

# Select the top 10 regions with the highest interest in "Machine Learning"
data = data.head(10)

# Print the resulting dataframe
print(data)
```

1. we first import the TrendReq class from the pytrends.request module. We then create an instance of the TrendReq class and use the build_payload() method to build a payload with the search term "Machine Learning". This tells the Google Trends API what data we are interested in retrieving.

2. Next, we use the interest_by_region() method to retrieve the interest by region data for the search term "Machine Learning". This data shows the relative popularity of the search term in different regions around the world.

3. We then sort the data by the "Machine Learning" column in descending order using the sort_values() method, which allows us to see which regions have the highest interest in the search term. We select the top 10 regions with the highest interest using the head() method and store the resulting dataframe in the data variable.

4. Finally, we print out the resulting dataframe using the print() function. The resulting dataframe will show the top 10 regions with the highest interest in the search term "Machine Learning", ranked in descending order.

```python
# Reset the index of the data frame
data = data.reset_index()

# Plot the data
data.plot(x="geoName", y="Machine Learning", kind="bar", figsize=(15, 12))

# Use the "fivethirtyeight" style for the plot
plt.style.use('fivethirtyeight')

# Show the plot
plt.show()
```

1. we first reset the index of the dataframe using the reset_index() method. This is because the interest_by_region() method returns a dataframe with the region names as the index, and we want to plot the data using the region names as a column.


2. We then use the plot() method to create a bar plot of the data. The x parameter is set to "geoName" to use the region names as the x-axis labels, and the y parameter is set to "Machine Learning" to plot the interest in the search term "Machine Learning" on the y-axis. We also set the kind parameter to "bar" to create a bar plot, and the figsize parameter to (15, 12) to set the size of the plot.


3. Next, we use the plt.style.use() method to set the plot style to "fivethirtyeight", which is a popular data visualization style.

4. Finally, we use the plt.show() method to display the plot. The resulting plot will show the top regions with the highest interest in the search term "Machine Learning" plotted as bars on the y-axis, with the region names displayed on the x-axis.


# let’s have a look at the trend of searches to see how the total search queries based on “Machine Learning” increased or decreased on Google:

```python
# Create a TrendReq object with the language and timezone parameters set
data = TrendReq(hl='en-US', tz=360)

# Build the payload with the search term "Machine Learning"
data.build_payload(kw_list=['Machine Learning'])

# Retrieve the interest over time data
data = data.interest_over_time()

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(15, 12))

# Plot the data for the "Machine Learning" search term
data['Machine Learning'].plot()

# Use the "fivethirtyeight" style for the plot
plt.style.use('fivethirtyeight')

# Set the title, x-label, and y-label for the plot
plt.title('Total Google Searches for Machine Learning', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')

# Show the plot
plt.show()
```

So we can see a huge increase in the searches about “machine learning” on Google in 2023. This is how we can analyze Google searches based on any keyword.
